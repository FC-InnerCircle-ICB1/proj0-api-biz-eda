```python
!pip install -q koreanize-matplotlib
import requests
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns
from itertools import product
from tqdm import tqdm
import time

# 네이버 API 인증 정보

from google.colab import userdata
CLIENT_ID = userdata.get('client_id')
CLIENT_SECRET = userdata.get('client_secret')

URL = "https://openapi.naver.com/v1/datalab/shopping/categories"

# API 요청 함수
def fetch_data(client_id, client_secret, url, categories, start_date, end_date, time_unit, genders, ages):
    headers = {
        "X-Naver-Client-Id": client_id,
        "X-Naver-Client-Secret": client_secret,
        "Content-Type": "application/json"
    }
    all_results = []

    # 모든 조합 생성
    combinations = list(product(categories, genders, ages))

    # tqdm으로 진행률 표시
    for category, gender, age in tqdm(combinations, desc="데이터 수집 중"):
        body = {
            "startDate": start_date,
            "endDate": end_date,
            "timeUnit": time_unit,
            "category": [category],
            "gender": gender,
            "ages": [age]
        }

        response = requests.post(url, headers=headers, json=body)
        if response.status_code == 200:
            data = response.json()
            for category_data in data["results"]:
                for data_point in category_data["data"]:
                    all_results.append({
                        "date": data_point["period"],
                        "category": category["name"],
                        "gender": gender,
                        "age": age,
                        "ratio": data_point["ratio"]
                    })
        else:
            print(f"API 요청 실패: {category['name']}, 상태 코드: {response.status_code}")

        # API 요청 제한 고려
        time.sleep(0.1)

    return pd.DataFrame(all_results)

# 향상된 데이터 시각화 함수
def visualize_data(df):
    # 날짜 변환
    df['date'] = pd.to_datetime(df['date'])
    df['month'] = df['date'].dt.month
    df['dayofweek'] = df['date'].dt.dayofweek


    # 1. 일자별 트렌드 (카테고리별)
    plt.figure(figsize=(15, 6))
    for category in df['category'].unique():
        category_data = df[df['category'] == category].groupby('date')['ratio'].mean()
        plt.plot(category_data.index, category_data.values, label=category, marker='o')
    plt.title('카테고리별 일별 트렌드')
    plt.xlabel('일자')
    plt.ylabel('평균 Ratio')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show()


    # 2. 연령대별 카테고리 선호도
    plt.figure(figsize=(12, 6))
    age_analysis = df.groupby(['category', 'age'])['ratio'].mean().unstack()
    sns.heatmap(age_analysis, annot=True, fmt='.1f', cmap='YlOrRd')
    plt.title('연령대별 카테고리 선호도')
    plt.xlabel('연령대')
    plt.ylabel('카테고리')
    plt.tight_layout()
    plt.show()

    # 3. 성별 카테고리 선호도
    plt.figure(figsize=(12, 6))
    gender_analysis = df.groupby(['category', 'gender'])['ratio'].mean().unstack()
    gender_analysis.plot(kind='bar', width=0.8)
    plt.title('성별 카테고리 선호도')
    plt.xlabel('카테고리')
    plt.ylabel('평균 Ratio')
    plt.xticks(rotation=45)
    plt.legend(title='성별')
    plt.tight_layout()
    plt.show()



# 메인 실행
if __name__ == "__main__":
    # 카테고리 설정
    CATEGORIES = [
        {"name": "패션의류", "param": ["50000000"]},
        {"name": "화장품/미용", "param": ["50000002"]},
        {"name": "디지털/가전", "param": ["50000003"]},
        {"name": "가구/인테리어", "param": ["50000004"]},
        {"name": "출산/육아", "param": ["50000005"]},
        {"name": "스포츠/레저", "param": ["50000006"]},
        {"name": "생활/건강", "param": ["50000007"]},
        {"name": "여가/생활편의", "param": ["50000008"]}
    ]

    # 데이터 수집 설정
    START_DATE = "2024-11-01"
    END_DATE = "2024-12-22"
    TIME_UNIT = "date"
    GENDERS = ["f", "m"]
    AGES = ["10", "20", "30", "40", "50", "60"]

    # 데이터 수집
    df = fetch_data(
        CLIENT_ID, CLIENT_SECRET, URL, CATEGORIES,
        START_DATE, END_DATE, TIME_UNIT,
        GENDERS, AGES
    )

    # 데이터 저장 및 시각화
    if not df.empty:
        print("\n데이터 수집 완료!")
        print(f"총 {len(df):,}개의 데이터 포인트 수집")

        # 기본 통계 출력
        print("\n=== 데이터 요약 ===")
        print(f"기간: {df['date'].min()} ~ {df['date'].max()}")
        print(f"카테고리 수: {len(df['category'].unique())}")
        print(f"수집된 성별: {', '.join(df['gender'].unique())}")
        print(f"수집된 연령대: {', '.join(df['age'].unique())}")

        # 데이터 저장
        df.to_csv("naver_shopping_data_full.csv", index=False)
        print("\n데이터가 'naver_shopping_data_full.csv'에 저장되었습니다.")

        # 데이터 시각화
        visualize_data(df)
    else:
        print("데이터 수집에 실패했습니다.")
```

```python
# 추가 데이터 분석 코드
import numpy as np
from sklearn.linear_model import LinearRegression

# 데이터 분석 함수
def analyze_and_plot_trends(df):
    plt.figure(figsize=(15, 6))

    categories = df['category'].unique()
    for category in categories:
        category_data = df[df['category'] == category].groupby('date')['ratio'].mean().reset_index()

        # 추세선 계산
        category_data['days'] = (category_data['date'] - category_data['date'].min()).dt.days
        X = category_data[['days']]
        y = category_data['ratio']
        model = LinearRegression()
        model.fit(X, y)
        trendline = model.predict(X)

        # 플롯 그리기
        plt.plot(category_data['date'], category_data['ratio'], label=f"{category} (데이터)", marker='o')
        plt.plot(category_data['date'], trendline, label=f"{category} (추세선)", linestyle='--')

    plt.title('카테고리별 일별 트렌드와 추세선')
    plt.xlabel('일자')
    plt.ylabel('평균 Ratio')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# 인사이트 도출 함수
def generate_insights(df):
    insights = []

    # 카테고리별 평균 변화율 계산
    for category in df['category'].unique():
        category_data = df[df['category'] == category].groupby('date')['ratio'].mean().reset_index()
        category_data['change'] = category_data['ratio'].pct_change() * 100
        avg_change = category_data['change'].mean()
        insights.append(f"카테고리 '{category}'의 평균 일별 변화율: {avg_change:.2f}%")

    # 성별/연령대별 가장 인기 있는 카테고리
    gender_age_popularity = df.groupby(['gender', 'age', 'category'])['ratio'].mean().idxmax()
    insights.append(f"성별 {gender_age_popularity[0]}, 연령대 {gender_age_popularity[1]}에서 가장 인기 있는 카테고리: {gender_age_popularity[2]}")

    return insights

# 추가 분석 아이디어
# 1. 특정 성별, 연령대에서 계절에 따른 카테고리 변화 분석
# 2. 상위 3개 카테고리의 주말과 주중 트렌드 비교
# 3. 연령대별 구매 트렌드 클러스터링

def suggest_additional_analysis():
    return [
        "연령대 그룹화에 따른 소비 패턴 클러스터링"
    ]

# 실행 코드
if not df.empty:
    print("\n=== 추가 데이터 분석 ===")

    # 트렌드 분석 및 시각화
    analyze_and_plot_trends(df)

    # 인사이트 도출
    insights = generate_insights(df)
    print("\n=== 데이터 인사이트 ===")
    for insight in insights:
        print(insight)

    # 추가 분석 제안
    suggestions = suggest_additional_analysis()
    print("\n=== 추가 분석 제안 ===")
    for suggestion in suggestions:
        print(f"- {suggestion}")
```
